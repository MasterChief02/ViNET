```python
import pandas as pd
```

```python
data_jio = pd.read_csv('website-jio.log')
```

```python
# assign every unique url to a unique id
url_id = {}
id_url = {}

def get_url_id(url):
    if url not in url_id:
        url_id[url] = len(url_id)
        id_url[url_id[url]] = url
    return url_id[url]

data_jio['url_id'] = data_jio['url'].apply(get_url_id)
```

```python
data_jio.boxplot(column=' time taken(s)', by='url_id', grid=False)
```
