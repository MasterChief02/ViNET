```python
import pandas as pd

data_jio = pd.read_csv('telegram-jio.log')
data_aitel = pd.read_csv('telegram-airtel.log')
```

```python
data_jio.columns
```

```python
# plot box plot of time_taken based on num_chars for jio and airtel

import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 2)
fig.set_size_inches(15, 5)

data_jio.boxplot(column=' time_taken', by='num_chars', ax=axs[0])
axs[0].set_title('Jio')
axs[0].set_ylabel('Time taken (s)')
axs[0].set_xlabel('Number of characters')

data_aitel.boxplot(column=' time_taken', by='num_chars', ax=axs[1])	
axs[1].set_title('Airtel')
axs[1].set_ylabel('Time taken (s)')
axs[1].set_xlabel('Number of characters')

plt.show()
```
