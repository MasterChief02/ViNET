```python
import numpy as np
import matplotlib.pyplot as plt
```

# 100K File Transfer

```python
with open("file_transfer_100k.log") as f:
    lines = f.readlines()

lines = lines[2:]
```

```python
lines = [line.split(',') for line in lines]
lines = [float(line[1].strip()) for line in lines]
lines
```

```python
print("Mean: ", np.mean(lines))
print("Median: ", np.median(lines))
print("Std Dev: ", np.std(lines))
```

```python
plt.plot(sorted(lines))
plt.show()
```

# 200k File Transfer

```python
with open("file_transfer_10_iters_200K.log") as f:
    output = f.readlines()

import re

# line starts with 100
pattern = r'^100.*$'

output = [re.findall(pattern, line) for line in output if re.findall(pattern, line)]
output = [line[0].split() for line in output]
output = [line[8] for line in output]
```

```python
import time

output = [time.strptime(line, "%H:%M:%S") for line in output]
output = [line.tm_hour*60*60 + line.tm_min*60 + line.tm_sec for line in output]
```

```python
print("Mean: ", np.mean(output))
print("Median: ", np.median(output))
print("Std Dev: ", np.std(output))
```

```python
plt.plot(output)
plt.show()
```
