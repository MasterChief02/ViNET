```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
from model_training import *
```

```python
ROOT_PATH = "/home/vinet/sec_analysis/forbidden_bit_60s/features"
dataset = Dataset(ROOT_PATH)
```

```python
vilte_pkt_time_in = np.zeros(9)
vinet_pkt_time_in = np.zeros(9)
vilte_pkt_time_out = np.zeros(9)
vinet_pkt_time_out = np.zeros(9)

df_vilte = dataset.c60.jio.vilte
df_vinet = dataset.c60.jio.vinet

for i in range(10, 100, 10):
    vilte_pkt_time_in[i // 10 - 1] = df_vilte[f"packet_time_in_p_{i}"].mean()
    vilte_pkt_time_out[i // 10 - 1] = df_vilte[f"packet_time_out_p_{i}"].mean()

    vinet_pkt_time_in[i // 10 - 1] = df_vinet[f"packet_time_in_p_{i}"].mean()
    vinet_pkt_time_out[i // 10 - 1] = df_vinet[f"packet_time_out_p_{i}"].mean()
```

```python
plt.plot(vilte_pkt_time_in * 1000, label="vilte_pkt_time_in")
plt.plot(vinet_pkt_time_in * 1000, label="vinet_pkt_time_in")
plt.legend()
plt.show()
```

```python
plt.plot(vilte_pkt_time_out * 1000, label="vilte_pkt_time_out")
plt.plot(vinet_pkt_time_out * 1000, label="vinet_pkt_time_out")
plt.legend()
plt.show()
```
