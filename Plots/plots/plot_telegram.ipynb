```python
import pandas as pd

data_jio = pd.read_csv('telegram-jio.log')
data_airtel = pd.read_csv('telegram-airtel.log')
```

```python
data_jio.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axs = plt.subplots(1, 2)
fig.set_size_inches(11, 3.3)
fig.suptitle('Telegram Speed Test')

# Define a common style for black lines
black_props = dict(color='black')

sns.boxplot(
    data=data_jio, 
    y=' time_taken', 
    x='num_chars', 
    ax=axs[0], 
    fill=False, 
    boxprops=black_props,
    medianprops=black_props,
    whiskerprops=black_props,
    capprops=black_props,
    flierprops={'markeredgecolor':'black'}
)
axs[0].set_title('Jio')
axs[0].set_ylabel('Time taken (s)')
axs[0].set_xlabel('Number of characters')
axs[0].grid(True)

sns.boxplot(
    data=data_airtel, 
    y=' time_taken', 
    x='num_chars', 
    ax=axs[1], 
    fill=False, 
    boxprops=black_props,
    medianprops=black_props,
    whiskerprops=black_props,
    capprops=black_props,
    flierprops={'markeredgecolor':'black'}
)
axs[1].set_title('Airtel')
axs[1].set_ylabel('Time taken (s)')
axs[1].set_xlabel('Number of characters')
axs[1].grid(True)

axs[0].set_xticklabels(['2', '4', '8', '16', '32', '64', '128', '256', '512', '1K', '2K', '4K'])
axs[1].set_xticklabels(['2', '4', '8', '16', '32', '64', '128', '256', '512', '1K', '2K', '4K'])

axs[0].set_ylim(0, 5)
axs[1].set_ylim(0, 5)

plt.show()

fig.savefig('telegram_speed.pdf', bbox_inches='tight')
```
