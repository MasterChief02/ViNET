```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from model_training import Dataset
```

# ROC AUC

```python
summ_curve_jio = np.loadtxt('numbers/Jio_ViLTE_vs_ViNET_roc_curve.txt', delimiter=',')
pl_curve_jio = np.loadtxt('numbers/Jio_ViLTE_vs_ViNET_pl_roc_curve.txt', delimiter=',')

with open("numbers/Jio_ViLTE_vs_ViNET_auc.txt") as f:
    lines = f.readlines()
    summ_auc_jio = float(lines[0].strip())
    chunk_summ_auc_jio = float(lines[1].strip())

with open("numbers/Jio_ViLTE_vs_ViNET_pl_auc.txt") as f:
    lines = f.readlines()
    pl_auc_jio = float(lines[0].strip())
    chunk_pl_auc_jio = float(lines[1].strip())
    

summ_curve_airtel = np.loadtxt('numbers/Airtel_ViLTE_vs_ViNET_roc_curve.txt', delimiter=',')
pl_curve_airtel = np.loadtxt('numbers/Airtel_ViLTE_vs_ViNET_pl_roc_curve.txt', delimiter=',')

with open("numbers/Airtel_ViLTE_vs_ViNET_auc.txt") as f:
    lines = f.readlines()
    summ_auc_airtel = float(lines[0].strip())
    chunk_summ_auc_airtel = float(lines[1].strip())
with open("numbers/Airtel_ViLTE_vs_ViNET_pl_auc.txt") as f:
    lines = f.readlines()
    pl_auc_airtel = float(lines[0].strip())
    chunk_pl_auc_airtel = float(lines[1].strip())
    
summ_curve_vi = np.loadtxt('numbers/VI_ViLTE_vs_VInet_roc_curve.txt', delimiter=',')
pl_curve_vi = np.loadtxt('numbers/VI_ViLTE_vs_VInet_pl_roc_curve.txt', delimiter=',')

with open("numbers/VI_ViLTE_vs_VInet_auc.txt") as f:
    lines = f.readlines()
    summ_auc_vi = float(lines[0].strip())
    chunk_summ_auc_vi = float(lines[1].strip())
with open("numbers/VI_ViLTE_vs_VInet_pl_auc.txt") as f:
    lines = f.readlines()
    pl_auc_vi = float(lines[0].strip())
    chunk_pl_auc_vi = float(lines[1].strip())
```

```python
pallete = sns.color_palette("rocket")
pallete = [pallete[2], pallete[3], pallete[5]]
```

```python
fig, axs = plt.subplots(1, 3, figsize=(11, 3.5))

axs[0].plot(summ_curve_airtel[:, 0], summ_curve_airtel[:, 1], label=f'Sum Stats AUC: {summ_auc_airtel:.2f}', lw=2, color=pallete[0])
axs[0].plot(pl_curve_airtel[:, 0], pl_curve_airtel[:, 1], label=f'Quant PLs AUC: {pl_auc_airtel:.2f}', lw=2, color=pallete[0], linestyle='-.')
axs[0].plot([0, 1], [0, 1], linestyle='--', lw=2, color='gray', label='Chance', alpha=.8)
axs[0].set_title(AIRTEL, fontsize=12)
axs[0].grid(axis='y')
axs[0].set_xlim(0, 1)
axs[0].set_ylim(0, 1)
axs[0].spines["top"].set_visible(False)
axs[0].spines["right"].set_visible(False)
axs[0].legend(fontsize='small')
# axs[0].set_ylabel('True Positive Rate', fontsize=12)
# axs[0].set_xlabel('False Positive Rate', fontsize=12)


axs[1].plot(summ_curve_jio[:, 0], summ_curve_jio[:, 1], label=f'Sum Stats AUC: {summ_auc_jio:.2f}', lw=2, color=pallete[1])
axs[1].plot(pl_curve_jio[:, 0], pl_curve_jio[:, 1], label=f'Quant PLs AUC: {pl_auc_jio:.2f}', lw=2, color=pallete[1], linestyle='-.')    
axs[1].plot([0, 1], [0, 1], linestyle='--', lw=2, color='gray', label='Chance', alpha=.8)
axs[1].set_title(JIO, fontsize=12)
axs[1].grid(axis='y')
axs[1].set_xlim(0, 1)
axs[1].set_ylim(0, 1)
axs[1].spines["top"].set_visible(False)
axs[1].spines["right"].set_visible(False)
axs[1].legend(fontsize='small')
# axs[1].set_ylabel('True Positive Rate', fontsize=12)
# axs[1].set_xlabel('False Positive Rate', fontsize=12)

axs[2].plot(summ_curve_vi[:, 0], summ_curve_vi[:, 1], label=f'Sum Stats AUC: {summ_auc_vi:.2f}', lw=2, color=pallete[2])
axs[2].plot(pl_curve_vi[:, 0], pl_curve_vi[:, 1], label=f'Quant PLs AUC: {pl_auc_vi:.2f}', lw=2, color=pallete[2], linestyle='-.')    
axs[2].plot([0, 1], [0, 1], linestyle='--', lw=2, color='gray', label='Chance', alpha=.8)
axs[2].set_title(VI, fontsize=12)
axs[2].grid(axis='y')
axs[2].set_xlim(0, 1)
axs[2].set_ylim(0, 1)
axs[2].spines["top"].set_visible(False)
axs[2].spines["right"].set_visible(False)
axs[2].legend(fontsize='small')
# axs[2].set_ylabel('True Positive Rate', fontsize=12)
# axs[2].set_xlabel('False Positive Rate', fontsize=12)

fig.text(0.5, 0.04, 'False Positive Rate', ha='center', va='center', fontsize=12)
fig.text(0.04, 0.5, 'True Positive Rate', ha='center', va='center', rotation='vertical', fontsize=12)

plt.tight_layout(rect=[0.05, 0.05, 1, 1])
# plt.xlabel('False Positive Rate', fontsize=12)
# plt.ylabel('True Positive Rate', fontsize=12)
plt.savefig(f"paper_figures/ViLTE_vs_ViNET_roc_curve_mno.pdf", bbox_inches='tight')
plt.show()

```

```python
summ_curve_jio_one_transfer = np.loadtxt('numbers/Jio_ViLTE_vs_ViNET_one_transfer_roc_curve.txt', delimiter=',')
pl_curve_jio_one_transfer = np.loadtxt('numbers/Jio_ViLTE_vs_ViNET_pl_one_transfer_roc_curve.txt', delimiter=',')
with open("numbers/Jio_ViLTE_vs_ViNET_one_transfer_auc.txt") as f:
    lines = f.readlines()
    summ_auc_jio_one_transfer = float(lines[0].strip())
    chunk_summ_auc_jio_one_transfer = float(lines[1].strip())

with open("numbers/Jio_ViLTE_vs_ViNET_pl_one_transfer_auc.txt") as f:
    lines = f.readlines()
    pl_auc_jio_one_transfer = float(lines[0].strip())
    chunk_pl_auc_jio_one_transfer = float(lines[1].strip())

```
